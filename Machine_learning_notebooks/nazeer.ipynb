```python
from collections import Counter
students_marks = []
with open('student_marks.csv') as f:
    for line in f:
        columns = line.split(',')
        columns[-1] = int(columns[-1].rstrip('\n'))
        students_marks.append(tuple(columns))
subject_faculty = []
with open('subject_faculty.csv') as f:
    for line in f:
        columns = line.split(',')
        columns[-1] = columns[-1].rstrip('\n')
        subject_faculty.append(tuple(columns))
#===================================================================

#1.Find the faculty with highest student count who got more than 90%. 
def highest_student_count():
    l = []
    for x in students_marks:
        if x[2] >= 90:
            l.append(x[1])
    counter_sub = Counter(l)
    #print(counter_sub)
    max_count_subject = max(counter_sub.items(),key=lambda x:x[1])
    #print(max_count_subject)
    for subject, faculty in subject_faculty:
        if subject == max_count_subject[0]:
            print('1.faculty with highest student count:',faculty)
            print()
highest_student_count()
#=====================================================================           
#2.Find the faculty with highest pass percentage (> 40%)
def highest_pass_percentage():
    l2 = []
    for p in students_marks:
        if p[2] >=40:
            l2.append(p[1])
    counter_pass_per = Counter(l2)
    #print(counter_pass_per)
    max_count_per = max(counter_pass_per.items(),key=lambda p:p[1])
    #print(max_count)
    for subject, faculty in subject_faculty:
        if subject == max_count_per[0]:
            print("2.faculty with highest pass percentage:",faculty)
            print()
highest_pass_percentage()
#======================================================================
#3.Find the faculty with least pass percentage (<= 40%)
def least_pass_percentage():
    l3 = []
    for least in students_marks:
        if least[2] <= 40:
            l3.append(least[1])
    count_least = Counter(l3)
    #print(count_least)
    min_least_per = max(count_least.items(),key = lambda least:least[1])
    #print(min_least_per)
    for subject, faculty in subject_faculty:
        if subject == min_least_per[0]:
            print('3.faculty with least pass percentage:',faculty)
            print()
least_pass_percentage()
#=========================================================================
#4.Who is the top student with maximum total?
def top_student_max_total():
    total_marks = {}
    for name, subject ,marks in students_marks:
        if name not in total_marks:
           total_marks[name] = marks
        else :
           total_marks[name] = total_marks[name] + marks
    max_marks = max(total_marks.items(),key=lambda x:x[1])
    print("4.Top_student_with_maximum_total:",max_marks)
top_student_max_total()
print()
#==========================================================================
#5.Who is the best student in Mathematics?
def best_student_in_Math():
    best_student = {}
    for name,sub,marks in students_marks:
        if sub == "Mathematics":
            best_student[name] = marks
    max_marks = max(best_student.items(),key=lambda x:x[1])
    print("5.the best student in Mathematics:",max_marks)
best_student_in_Math()
print()
#==========================================================================
#6.What is the average mark for each subject,(ignore failures)?
from statistics import mean

def avg_marks_each_subject():
    from statistics import mean

    subject_wise = {}
    for name,subject,marks in students_marks:
        if marks >= 40:
            if subject not in subject_wise:
                subject_wise[subject] = [marks]
            else:
                subject_wise[subject].append(marks)
    #print("list of subject-wise marks:",subject_wise)
    total_sub_marks= {k:sum(v) for k,v in subject_wise.items()}
    print("6.Total marks subject-wise",total_sub_marks)
    print()
    for k, v in subject_wise.items():
        d = k,mean(v)
        d2 = tuple(map(lambda x: isinstance(x, float) and round(x, 2) or x, d))
        print("average mark for each subject:",d2)
        print()
avg_marks_each_subject()
#==========================================================================
#7.Find the student with least numbers of marks as total.
def leat_student_marks_total():
    total_marks = {}
    for name, subject ,marks in students_marks:
        if name not in total_marks:
           total_marks[name] = marks
        else :
           total_marks[name] = total_marks[name] + marks
    least_marks = min(total_marks.items(),key=lambda x:x[1])
    print("7.least_student_marks:",least_marks)
leat_student_marks_total()
#===========================================================================
        

```
