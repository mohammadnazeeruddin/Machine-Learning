```python
import pandas as pd
import numpy as np
```

```python
mobiles_info=pd.read_csv('mobiles_info_dataset.csv')
```

```python
mobiles_info.head()
```

```python
mobiles_info.columns
```

```python
from sklearn.preprocessing import LabelEncoder

price_lb = LabelEncoder()
ram_lb = LabelEncoder()
Rom_lb = LabelEncoder()
Expandable_lb = LabelEncoder()
Screen_in_inch_lb = LabelEncoder()
Secondary_Camera_lb = LabelEncoder()
Battery_details_lb= LabelEncoder()
Warranty_lb= LabelEncoder()

mobiles_info['Price in Rupees'] = price_lb.fit_transform(mobiles_info['Price in Rupees'])
mobiles_info['Ram'] = ram_lb.fit_transform(mobiles_info.Ram)
mobiles_info['Rom'] = Rom_lb.fit_transform(mobiles_info.Rom)
mobiles_info['Expandable_GB/TB'] = Expandable_lb.fit_transform(mobiles_info['Expandable_GB/TB'])
mobiles_info['Screen_in_inch'] = Screen_in_inch_lb.fit_transform(mobiles_info.Screen_in_inch)
mobiles_info['Secondary_Camera(MP)'] = Secondary_Camera_lb.fit_transform(mobiles_info['Secondary_Camera(MP)'])
mobiles_info['Battery_details'] = Battery_details_lb.fit_transform(mobiles_info.Battery_details)
mobiles_info['Warranty(Year)'] = Warranty_lb.fit_transform(mobiles_info['Warranty(Year)'])
```

```python
from sklearn.preprocessing import OneHotEncoder

price_lb = OneHotEncoder(handle_unknown='ignore')
ram_lb = OneHotEncoder(handle_unknown='ignore')
Rom_lb = OneHotEncoder(handle_unknown='ignore')
Expandable_lb = OneHotEncoder(handle_unknown='ignore')
Screen_in_inch_lb = OneHotEncoder(handle_unknown='ignore')
Secondary_Camera_lb = OneHotEncoder(handle_unknown='ignore')
Battery_details_lb= OneHotEncoder(handle_unknown='ignore')
Warranty_lb= OneHotEncoder(handle_unknown='ignore')

mobiles_info['Price in Rupees'] =price_lb.fit_transform(mobiles_info['Price in Rupees'].values.reshape(-1,1)).toarray()
mobiles_info['Ram'] = ram_lb.fit_transform(mobiles_info.Ram.values.reshape(-1,1)).toarray()
mobiles_info['Rom'] = Rom_lb.fit_transform(mobiles_info.Rom.values.reshape(-1,1)).toarray()
mobiles_info['Expandable_GB/TB'] = Expandable_lb.fit_transform(mobiles_info['Expandable_GB/TB'].values.reshape(-1,1)).toarray()
mobiles_info['Screen_in_inch'] = Screen_in_inch_lb.fit_transform(mobiles_info.Screen_in_inch.values.reshape(-1,1)).toarray()
mobiles_info['Secondary_Camera(MP)'] = Secondary_Camera_lb.fit_transform(mobiles_info['Secondary_Camera(MP)'].values.reshape(-1,1)).toarray()
mobiles_info['Battery_details'] = Battery_details_lb.fit_transform(mobiles_info.Battery_details.values.reshape(-1,1)).toarray()
mobiles_info['Warranty(Year)'] = Warranty_lb.fit_transform(mobiles_info['Warranty(Year)'].values.reshape(-1,1)).toarray()
```
