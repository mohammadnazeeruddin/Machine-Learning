```python
import pandas as pd
import numpy as np
# import pandas_profiling as pp
```

```python
a_junction_lanes = pd.DataFrame(np.random.randint(5,20,size=(15000, 6)), columns=['a_lane1_input','a_lane2_input','a_lane1_output','a_lane2_output','a_lane3_output','a_lane4_output'])
```

```python
b_junction_lanes = pd.DataFrame(np.random.randint(5,20,size=(15000, 6)), columns=['b_lane1_input','b_lane2_input','b_lane1_output','b_lane2_output','b_lane3_output','b_lane4_output'])
```

```python
c_junction_lanes = pd.DataFrame(np.random.randint(5,20,size=(15000, 6)), columns=['c_lane1_input','c_lane2_input','c_lane1_output','c_lane2_output','c_lane3_output','c_lane4_output'])
```

```python
d_junction_lanes = pd.DataFrame(np.random.randint(5,20,size=(15000, 8)), columns=['d_lane1_input','d_lane2_input','d_lane3_input','d_lane1_output','d_lane2_output','d_lane3_output','d_lane4_output','d_lane5_output'])
```

```python
# A junction block
```

```python
# B junction vehicles (B junction sum after spliting vehicles) count added to the A junction 
# D junction vehicles (D junction sum after spliting vehicles) count added to the A junction 

a_junction_lanes['abd_lane_input_sum'] = a_junction_lanes['a_lane1_input'] + a_junction_lanes['a_lane2_input'] + b_junction_lanes['b_lane1_input'] + d_junction_lanes['d_lane1_input']
```

```python
# A junction output lanes from four output lanes

a_junction_lanes['a_jun_output_sum'] = a_junction_lanes['a_lane1_output'] + a_junction_lanes['a_lane2_output']+ a_junction_lanes['a_lane3_output']+ a_junction_lanes['a_lane4_output']
```

```python
# B junction block
```

```python
# A junction vehicles (A junction sum after spliting vehicles) count added to the B junction 
# C junction vehicles (C junction sum after spliting vehicles) count added to the B junction 


b_junction_lanes['bac_lane_input_sum'] = b_junction_lanes['b_lane1_input'] + b_junction_lanes['b_lane2_input'] + a_junction_lanes['a_lane1_input'] + c_junction_lanes['c_lane1_input']
```

```python
# B junction output lanes from four output lanes

b_junction_lanes['b_jun_output_sum'] = b_junction_lanes['b_lane1_output'] + b_junction_lanes['b_lane2_output']+ b_junction_lanes['b_lane3_output']+ b_junction_lanes['b_lane4_output']
```

```python
b_junction_lanes
```

```python
# C junction block
```

```python
# B junction vehicles (B junction sum after spliting vehicles) count added to the C junction 
# D junction vehicles (D junction sum after spliting vehicles) count added to the C junction 

c_junction_lanes['cbd_lane_input_sum'] = c_junction_lanes['c_lane1_input'] + c_junction_lanes['c_lane2_input'] + b_junction_lanes['b_lane1_input'] + d_junction_lanes['d_lane1_input'] 
```

```python
# C junction output lanes from four output lanes

c_junction_lanes['c_jun_output_sum'] = c_junction_lanes['c_lane1_output'] + c_junction_lanes['c_lane2_output']+ c_junction_lanes['c_lane3_output']+ c_junction_lanes['c_lane4_output']
```

```python
c_junction_lanes 
```

```python
# D junction block
```

```python
# A junction vehicles (A junction sum after spliting vehicles) count added to the D junction 
# C junction vehicles (C junction sum after spliting vehicles) count added to the D junction 

d_junction_lanes['dac_lane_input_sum'] = d_junction_lanes['d_lane1_input'] + d_junction_lanes['d_lane2_input']+ d_junction_lanes['d_lane3_input'] + a_junction_lanes['a_lane1_input'] + c_junction_lanes['c_lane1_input'] 
```

```python
# D junction output lanes from four output lanes

d_junction_lanes['d_jun_output_sum'] = d_junction_lanes['d_lane1_output'] + d_junction_lanes['d_lane2_output']+ d_junction_lanes['d_lane3_output']+ d_junction_lanes['d_lane4_output']+d_junction_lanes['d_lane5_output']
```

```python
d_junction_lanes 
```

```python
result = pd.concat([a_junction_lanes,b_junction_lanes,c_junction_lanes,d_junction_lanes], axis=1)
```

```python
result.to_csv("traffic_junctions_dataset.csv")
```
