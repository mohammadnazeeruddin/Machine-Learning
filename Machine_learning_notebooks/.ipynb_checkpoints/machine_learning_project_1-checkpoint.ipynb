```python
# python required packages for data analysis.
import pandas as pd 
import pandas_profiling as pp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading csv file using pandas method read_csv().
mobiles_data = pd.read_csv('Flipkart_mobiles_data700.csv')
#mobiles_data is a dataframe(df)
```

```python
mobiles_data.drop_duplicates(subset='Mobiles', keep="first")
mobiles_data.drop_duplicates(subset='Colors', keep="first")
```

```python
mobiles_data.loc[mobiles_data.duplicated(keep='last'),:]
```

```python
# Removing duplicate values from Mobiles column.
mobiles_data =mobiles_data.groupby(['Colors']).size().reset_index()
# mobiles_data.rename(columns = {0: 'frequency'}, inplace = True)
```

```python
mobiles_data
```

```python
#data profiling using pandas_profiling lib
pp.ProfileReport(mobiles_data)
# The first section is an overview of the dataset and a list of possible issues with the data.
# The next section gives a summary of each column. You can click "toggle details" for even more information.
# The third section shows a heatmap of the correlation between columns.
# And the fourth section shows the head of the dataset.
```

```python
# data of all mobiles.
mobiles_data.head() 
```

```python
# we have 695 rows for each columns.
mobiles_data.tail()
```

```python
#droping mobiles,colors & Display columns using pandas.
mobiles_data.drop(['Mobiles','Colors','Display','Off(%)','Rating','Primary_Camera(MP)'], axis='columns', inplace=True)
```

```python
# dropped four columns Mobiles,colors,Display,offer
mobiles_data.head()
```

```python
mobiles_data.columns
```

```python
len(mobiles_data.columns)
```

```python
#handling errors
import re
# mobiles_data['Ram']=mobiles_data['Ram'].str.extract('(\d+)').head()
mobiles_data['Ram'] = mobiles_data['Ram'].apply(lambda x: re.search(r'\d+', x).group())
```

```python
mobiles_data['Expandable_GB/TB'] = mobiles_data['Expandable_GB/TB'].apply(lambda y: re.search(r'\d+', y).group())
```

```python
mobiles_data['Rom'] = mobiles_data['Rom'].str.replace('GB','')
```

```python
mobiles_data['Dual_sim'] = mobiles_data['Dual_sim'].str.replace('yes','1')
```

```python
# to find nan vaules
mobiles_data.loc[:, mobiles_data.isnull().any()].head()
```

```python
import math
```

```python
Rom_median=math.floor(mobiles_data.Rom.mode())
```

```python
Rom_median
```

```python
#handling missing values (NaN values)
mobiles_data['Rom'] = mobiles_data.Rom.fillna(Rom_median)
```

```python
mobiles_data.head(20)
```

```python
#finding agian  nan vaules
mobiles_data.loc[:, mobiles_data.isnull().any()].head(2) 
# no NaN values,now we have clean data set.
```

```python
# saved data set by using pandas method to_csv()
mobiles_data.to_csv('clean_mobiles_dataset.csv')
```

```python
# Linear Regression With One Variable
```

```python
plt.xlabel('Ram')
plt.ylabel('price')
plt.scatter(mobiles_data['Ram'],mobiles_data['Price in Rupees'],color='red',marker='+')
```

```python
from sklearn import linear_model
```

```python
# Create linear regression object
lr_model = linear_model.LinearRegression()
lr_model.fit(mobiles_data[['Ram']],mobiles_data['Price in Rupees'])
```

```python
# predicting price based on Ram feature.
prediction_price=lr_model.predict([[4]])  
```

```python
prediction_price
```

```python
lr_model.coef_
```
